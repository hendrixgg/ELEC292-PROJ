```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns=['A', 'B'], index=np.arange(0.1, 10, 0.1))

df.index = pd.to_datetime(df.index, unit='s')
```

```python
print(df)
```

```python
i= 0
intervals = []
for interval in df.rolling(window='1.1s'):
    print()
    if interval.index[-1] - interval.index[0] < pd.Timedelta('1s'):
        continue
    print(interval)
    intervals.append(interval)
```

```python
print(intervals[i].reset_index().iloc[10, 0].second + intervals[i].reset_index().iloc[10, 0].microsecond/1e6)
interval = intervals[i].reset_index()
interval['index'] = interval['index'].apply(lambda x: x.second + x.microsecond/1e6)
print(interval)
```

```python

print([x.value / 1e9 for x in intervals[i].reset_index()['index']])
```
